# Application Case on a P2P event log

```python
from semconstmining.parsing.label_parser.nlp_helper import NlpHelper
import pandas as pd
from semconstmining.main import get_resource_handler, get_or_mine_constraints, compute_relevance_for_log, get_parts_of_constraints, recommend_constraints_for_log
from semconstmining.config import Config
from pathlib import Path
import os
```

```python
conf = Config(Path(os.getcwd()).parents[0].resolve(), "opal")
nlp_helper = NlpHelper(conf)
rh = get_resource_handler(config=conf, nlp_helper=nlp_helper)
pd.set_option('display.max_rows', 10000)
```

```python
consts = get_or_mine_constraints(conf, rh)
nlp_helper.pre_compute_embeddings(sentences=get_parts_of_constraints(conf, consts))
```

```python
pd.set_option('display.min_rows', 20000)
consts[(consts["Level"]=="Object") & (consts["op_type"]=="Binary") & (consts["left_op"]=="create") & (consts["right_op"]=="reject")]
```

```python
CURRENT_LOG_FILE = "p2p_test.xes"

consts_with_relevance=compute_relevance_for_log(conf, consts, nlp_helper, CURRENT_LOG_FILE, precompute=True)
```

```python
consts_with_relevance
```

```python
from semconstmining.selection.instantiation.recommendation_config import RecommendationConfig
rec_config = RecommendationConfig(conf)
recommended_constraints = recommend_constraints_for_log(conf, rec_config, consts_with_relevance, nlp_helper,
                                                            CURRENT_LOG_FILE)
```

```python
recommended_constraints[(recommended_constraints["Level"]=="Object") & (recommended_constraints["op_type"]=="Binary")]
#recommended_constraints.loc["0c7e02b2-e94f-4e66-85c4-6be8af5426b0_Object_purchase order"]
```
